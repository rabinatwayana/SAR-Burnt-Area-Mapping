```python
import rasterio
import matplotlib.pyplot as plt
import numpy as np
import glob
import os

# Set directory path containing TIFF images
tiff_dir = "output/prediction"

# Find all TIFF files in the directory
tiff_files = glob.glob(os.path.join(tiff_dir, "*.tif"))

# Loop through each TIFF file and display it
for tiff_path in tiff_files:
    with rasterio.open(tiff_path) as src:
        image = src.read(1)  # Read the first band
        profile = src.profile  # Get metadata

    # Display image properties
    print(f"TIFF Metadata for {os.path.basename(tiff_path)}:", profile)

    # Plot the first band
    plt.figure(figsize=(10, 6))
    plt.imshow(image, cmap="gray")  # Change cmap as needed
    plt.colorbar(label="Pixel Values")
    plt.title(f"TIFF Image - {os.path.basename(tiff_path)}")
    plt.axis("off")
    plt.show()
```
